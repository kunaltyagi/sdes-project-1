```python
%matplotlib inline
%run oscillator.py
%run plot.py
from matplotlib import animation, rc
from IPython.display import HTML
```

```python
a = Oscillator(mu=2, x0=2)
ans = a.solve(0, 24, 0.01)
x1 = Sequence(ans[:,0], '$x$')
x2 = Sequence(ans[:,1], '$\dot{x}$')
```

```python
fig=plt.figure()
ax=fig.add_subplot(111,autoscale_on=False, xlim=(0,6), ylim=(-4,4))
ax.grid()
line1, line2, = ax.plot([],[],'b-',[],[],'r-', lw=2)
time_template = 'time = %.1fs'
time_text = ax.text(0.05, 0.9, '', transform=ax.transAxes)
```

```python
def init():
    line1.set_data([], [])
    line2.set_data([], [])
    time_text.set_text('')
    return line, time_text
```

```python
def animate_line(i):
    y = np.arange(0,6, 0.01)
    line1.set_data(y, x1.data[i:600+i])
    line2.set_data(y, x2.data[i:600+i])
    time_text.set_text(time_template % (i*0.01))
    return line, time_text
```

```python
x1.plot()
plt.show()
```

```python
anim = animation.FuncAnimation(fig, animate_line, init_func=init,
                               frames=760*2, interval=20, blit=True)
```

```python
HTML(anim.to_html5_video())
```

```python
def animate_point(i):
    i = i*2
    line1.set_data(x1.data[i], x2.data[i])
    line2.set_data(x1.data[0:i], x2.data[0:i])
    time_text.set_text(time_template % (i*0.01))
    return line, time_text
```

```python
anim = animation.FuncAnimation(fig, animate_point, init_func=init,
                               frames=760, interval=20, blit=True)
HTML(anim.to_html5_video())
```
